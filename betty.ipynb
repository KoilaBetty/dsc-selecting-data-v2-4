CONNECTING TO SQLITE DATABASE USING PYTHON

```python
import sqlite3
conn = sqlite3.connect('data.sqlite')
```

SELECTING EMPLOYEES TABLE FROM THE DATABASE

```python
import pandas as pd
pd.read_sql("""select *FROM employees;""",conn)
```

```python
pd.read_sql("""
SELECT *
  FROM employees;
""", conn)
```

SELECTING SPECIFIC COLUMNS 

```python
pd.read_sql("""
SELECT lastName,firstName,email
FROM employees;
""",conn).head()
```

CHANGE COLUMN NAME

```python
pd.read_sql("""
SELECT lastName AS LastName,firstName AS FirstName
FROM employees;
""",conn).head()
```

CASE TO BIN COLUMN NAME

```python
pd.read_sql(""" 
SELECT firstName,lastName,jobTitle,
            CASE 
            WHEN jobTitle = "Sales Rep" THEN "Sales Rep"
            ELSE "Not Sales Rep"
            END AS Role
            FROM employees
""",conn).head(10)
```

CASE to Make Values Human-Readable

```python
pd.read_sql("""
SELECT firstName, lastName, officeCode,
       CASE
       WHEN officeCode = "1" THEN "San Francisco, CA"
       WHEN officeCode = "2" THEN "Boston, MA"
       WHEN officeCode = "3" THEN "New York, NY"
       WHEN officeCode = "4" THEN "Paris, France"
       WHEN officeCode = "5" THEN "Tokyo, Japan"
       END AS Office
  FROM employees;
""", conn).head(10)
```

Shorter syntax possible

```python
pd.read_sql("""
SELECT firstName, lastName, officeCode,
       CASE officeCode
       WHEN "1" THEN "San Francisco, CA"
       WHEN "2" THEN "Boston, MA"
       WHEN "3" THEN "New York, NY"
       WHEN "4" THEN "Paris, France"
       WHEN "5" THEN "Tokyo, Japan"
       END AS Office
  FROM employees;
""", conn).head(10)
```

Combine multiple column values

```python
pd.read_sql("""
SELECT firstName || lastName AS full_name
  FROM employees;
""", conn).head()
```

Concatenate those column values with a space (" ") string literal:

```python
pd.read_sql("""
SELECT firstName || " " || lastName AS full_name
  FROM employees;
""", conn).head()
```

**Built-in SQL Functions for Math Operations**

```python
pd.read_sql("""SELECT * FROM orderDetails;""", conn)
```

*Round*

```python
pd.read_sql("""
SELECT round(priceEach) As rounded_Price
FROM orderDetails;
""",conn)
```

*CAST*

```python
pd.read_sql("""
SELECT CAST (round(priceEach)As INTEGER) As rounded_price_int
FROM orderDetails;
""",conn)
```

how many days there are between the requiredDate and the orderDate for each order

```python
pd.read_sql("""
SELECT julianday(requiredDate)-julianday(orderDate) As days_from_order_to_required
FROM orders;
""",conn)
```
